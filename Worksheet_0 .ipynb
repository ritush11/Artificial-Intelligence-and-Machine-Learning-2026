```python
#4.1 Exercise on Functions:

#TASK 1 – Unit Converter Program

def unit_converter():
    """
    Converts between different units of measurement.

    Conversion Types:
    1. Length (meters ↔ feet)
    2. Weight (kg ↔ lbs)
    3. Volume (liters ↔ gallons)

    Returns:
        float: Converted value
    """

    print("Choose conversion type:")
    print("1. Length")
    print("2. Weight")
    print("3. Volume")

    choice = input("Enter choice (1/2/3): ")

    try:
        value = float(input("Enter value to convert: "))

        if choice == "1":
            print("a. meters to feet")
            print("b. feet to meters")
            sub = input("Choose option (a/b): ")

            if sub == "a":
                result = value * 3.28084
                print(f"{value} meters = {result} feet")
                return result
            elif sub == "b":
                result = value / 3.28084
                print(f"{value} feet = {result} meters")
                return result

        elif choice == "2":
            print("a. kg to lbs")
            print("b. lbs to kg")
            sub = input("Choose option (a/b): ")

            if sub == "a":
                result = value * 2.20462
                print(f"{value} kg = {result} lbs")
                return result
            elif sub == "b":
                result = value / 2.20462
                print(f"{value} lbs = {result} kg")
                return result

        elif choice == "3":
            print("a. liters to gallons")
            print("b. gallons to liters")
            sub = input("Choose option (a/b): ")

            if sub == "a":
                result = value * 0.264172
                print(f"{value} liters = {result} gallons")
                return result
            elif sub == "b":
                result = value / 0.264172
                print(f"{value} gallons = {result} liters")
                return result

        else:
            print("Invalid conversion type.")
            return None

    except ValueError:
        print("Invalid input! Please enter numeric value.")
        return None


# Calling the function
unit_converter()
```

```python
#Task 2: Mathematical Operations on List

def sum_numbers(numbers):
    """
    Returns the sum of a list of numbers.
    """
    return sum(numbers)


def average_numbers(numbers):
    """
    Returns the average of a list of numbers.
    """
    return sum(numbers) / len(numbers)


def max_number(numbers):
    """
    Returns the maximum value from a list.
    """
    return max(numbers)


def min_number(numbers):
    """
    Returns the minimum value from a list.
    """
    return min(numbers)


def math_operations():
    """
    Performs selected mathematical operation on user-provided list.
    """

    print("Choose operation:")
    print("1. Sum")
    print("2. Average")
    print("3. Maximum")
    print("4. Minimum")

    choice = input("Enter choice (1/2/3/4): ")

    try:
        numbers = list(map(float, input("Enter numbers separated by spaces: ").split()))

        if len(numbers) == 0:
            print("Empty list not allowed.")
            return

        if choice == "1":
            print("Sum:", sum_numbers(numbers))
        elif choice == "2":
            print("Average:", average_numbers(numbers))
        elif choice == "3":
            print("Maximum:", max_number(numbers))
        elif choice == "4":
            print("Minimum:", min_number(numbers))
        else:
            print("Invalid operation choice.")

    except ValueError:
        print("Invalid input! Please enter numbers only.")


# Call the function
math_operations()
```

```python
#4.2 Exercise on List Manipulation:

# 1️ - Extract Every Other Element

def extract_every_other(lst):
    """
    Returns every other element from the list,
    starting from the first element.

    Parameters:
        lst (list): Input list

    Returns:
        list: New list containing every second element
    """
    return lst[::2]


# Example
print("1.", extract_every_other([1, 2, 3, 4, 5, 6]))
```

```python
# 2️ - Slice a Sublist

def get_sublist(lst, start, end):
    """
    Returns a sublist from index 'start' to 'end' (inclusive).

    Parameters:
        lst (list): Input list
        start (int): Starting index
        end (int): Ending index (inclusive)

    Returns:
        list: Sublist from start to end
    """
    return lst[start:end+1]


# Example
print("2.", get_sublist([1, 2, 3, 4, 5, 6], 2, 4))
```

```python
# 3️ - Reverse a List Using Slicing

def reverse_list(lst):
    """
    Returns the reversed version of the input list using slicing.

    Parameters:
        lst (list): Input list

    Returns:
        list: Reversed list
    """
    return lst[::-1]


# Example
print("3.", reverse_list([1, 2, 3, 4, 5]))
```

```python
# 4️ - Remove the First and Last Elements

def remove_first_last(lst):
    """
    Removes the first and last elements from the list.

    Parameters:
        lst (list): Input list

    Returns:
        list: List without first and last elements
    """
    return lst[1:-1]


# Example
print("4.", remove_first_last([1, 2, 3, 4, 5]))
```

```python
# 5️ - Get the First n Elements

def get_first_n(lst, n):
    """
    Returns the first n elements of the list.

    Parameters:
        lst (list): Input list
        n (int): Number of elements to extract

    Returns:
        list: First n elements
    """
    return lst[:n]


# Example
print("5.", get_first_n([1, 2, 3, 4, 5], 3))
```

```python
# 6 -  Extract the Last n Elements

def get_last_n(lst, n):
    """
    Returns the last n elements of the list.

    Parameters:
        lst (list): Input list
        n (int): Number of elements to extract

    Returns:
        list: Last n elements
    """
    return lst[-n:]


# Example
print("6.", get_last_n([1, 2, 3, 4, 5], 2))
```

```python
# 7️ - Extract Elements in Reverse Order (Skip One)

def reverse_skip(lst):
    """
    Returns every second element starting from
    the second-to-last element, moving backward.

    Parameters:
        lst (list): Input list

    Returns:
        list: New list of reversed skipped elements
    """
    return lst[-2::-2]


# Example
print("7.", reverse_skip([1, 2, 3, 4, 5, 6]))
```

```python
# 4.3 Exercise on Nested List
# 1️ - Flatten a Nested List

def flatten(lst):
    """
    Flattens a nested list into a single list.

    Parameters:
        lst (list): Nested list

    Returns:
        list: Flattened list
    """
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(flatten(item))
        else:
            result.append(item)
    return result


# Example
print("1.", flatten([[1, 2], [3, 4], [5]]))
```

```python
# 2 - Accessing Nested List Elements

def access_nested_element(lst, indices):
    """
    Accesses an element in a nested list using a list of indices.

    Parameters:
        lst (list): Nested list
        indices (list): List of indices

    Returns:
        element at specified position
    """
    element = lst
    for index in indices:
        element = element[index]
    return element


# Example
lst_example = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print("2.", access_nested_element(lst_example, [1, 2]))
```

```python
# 3️ - Sum of All Elements in a Nested List

def sum_nested(lst):
    """
    Returns the sum of all numeric elements
    in a nested list (any depth).
    """
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total


# Example
print("3.", sum_nested([[1, 2], [3, [4, 5]], 6]))
```

```python
# 4️ - Remove Specific Element from a Nested List

def remove_element(lst, elem):
    """
    Removes all occurrences of elem
    from a nested list.
    """
    result = []
    for item in lst:
        if isinstance(item, list):
            result.append(remove_element(item, elem))
        elif item != elem:
            result.append(item)
    return result


# Example
print("4.", remove_element([[1, 2], [3, 2], [4, 5]], 2))
```

```python
# 5️ - Find the Maximum Element in a Nested List

def find_max(lst):
    """
    Returns the maximum element
    in a nested list.
    """
    flat_list = flatten(lst)
    return max(flat_list)


# Example
print("5.", find_max([[1, 2], [3, [4, 5]], 6]))
```

```python
# 6️ - Count Occurrences of an Element

def count_occurrences(lst, elem):
    """
    Counts how many times elem appears
    in a nested list.
    """
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1
    return count


# Example
print("6.", count_occurrences([[1, 2], [2, 3], [2, 4]], 2))
```

```python
# 7️ - Deep Flatten (List of Lists of Lists)

def deep_flatten(lst):
    """
    Flattens a deeply nested list
    into a single list.
    """
    return flatten(lst)


# Example
print("7.", deep_flatten([[[1, 2], [3, 4]], [[5, 6], [7, 8]]]))
```

```python
# 8️ - Nested List Average

def average_nested(lst):
    """
    Calculates the average of all elements
    in a nested list.
    """
    flat_list = flatten(lst)
    return sum(flat_list) / len(flat_list)


# Example
print("8.", average_nested([[1, 2], [3, 4], [5, 6]]))
```

```python
#Problem - 4: Matrix Operations:

#Given Matrices
A = np.array([[3, 4],
              [7, 8]])

B = np.array([[5, 3],
              [2, 1]])

print("Matrix A:\n", A)
print("\nMatrix B:\n", B)


#Prove A · A⁻¹ = I
A_inv = np.linalg.inv(A)
identity_result = np.dot(A, A_inv)

print("\nA Inverse:\n", A_inv)
print("\nA · A⁻¹:\n", identity_result)


#Prove AB ≠ BA
AB = np.dot(A, B)
BA = np.dot(B, A)

print("\nA · B:\n", AB)
print("\nB · A:\n", BA)



#Prove (AB)ᵀ = Bᵀ Aᵀ
AB_transpose = AB.T
BT_AT = np.dot(B.T, A.T)

print("\n(AB)^T:\n", AB_transpose)
print("\nB^T · A^T:\n", BT_AT)
```

```python
#Step 1: Represent in Matrix Form AX = B

A_sys = np.array([[2, -3, 1],
                  [1, -1, 2],
                  [3,  1, -1]])

B_sys = np.array([-1, -3, 9])

print("Matrix A:\n", A_sys)
print("\nMatrix B:\n", B_sys)


#Step 2: Solve using Inverse Method

A_sys_inv = np.linalg.inv(A_sys)
X = np.dot(A_sys_inv, B_sys)

print("\nSolution using Inverse Method:")
print("x =", X[0])
print("y =", X[1])
print("z =", X[2])


#Using np.linalg.inv (Directly)
X_direct = np.linalg.inv(A_sys).dot(B_sys)

print("\nSolution using np.linalg.inv():")
print(X_direct)
```

```python
# 10.2 Experiment: How Fast is NumPy?

import numpy as np
import time

#1. Element-wise Addition:

#Using Python Lists
list1 = list(range(1_000_000))
list2 = list(range(1_000_000))

start_time = time.time()
result_list_add = [list1[i] + list2[i] for i in range(len(list1))]
end_time = time.time()
print("Python List Addition Time:", end_time - start_time, "seconds")


#Using NumPy Arrays
array1 = np.array(list1)
array2 = np.array(list2)

start_time = time.time()
result_array_add = array1 + array2
end_time = time.time()
print("NumPy Array Addition Time:", end_time - start_time, "seconds")



#2. Element-wise Multiplication
#Using Python Lists
start_time = time.time()
result_list_mul = [list1[i] * list2[i] for i in range(len(list1))]
end_time = time.time()
print("Python List Multiplication Time:", end_time - start_time, "seconds")


#Using NumPy Arrays
start_time = time.time()
result_array_mul = array1 * array2
end_time = time.time()
print("NumPy Array Multiplication Time:", end_time - start_time, "seconds")
```
